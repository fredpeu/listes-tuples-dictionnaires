## CALCUL D´UN MOYENNE                             

A la fin du semestre, il est temps de calculer les moyennes de français avant l´édition des bulletins. Pronote exporte les notes de Marta et Tiago sous la forme d´une liste: 

```python
marta= [10, 15, 11, 18, 7, 12, 13, 16, 18]
tiago= [8, 11, 12, 14, 10, 9, 15, 12, 11,13,18]
```

<br>*Compléter la fonction `moyenne(liste)` qui permet de calculer la moyenne d´un élève arrondie au dixième.*

```python
# code
def moyenne(liste):
    somme=0
    for i in range(len(liste)):
        somme+=liste[i]
    return round(somme/len(liste),1)
```

<br>Affichez les moyennes de Marta et Tiago:

```python
# code
moyenne (marta)
```

<br>*Tiago a obtenu un **14** et Marta un **16***. Comment ajouter ces deux nouvelles notes aux listes?

```python
# code
moyenne(tiago)
```

<br>*Le professeur décide de cocher la case "enlever la plus mauvaise note". A l´aide de la fonction `indexMini` (à compléter), affichez la moyenne corrigée de Marta et Tiago.*

http://www.fredpeuriere.com/NSI/algos-1/algorithmes-illustres/min.html

```python
# on trouve d´abord l´indice de la plus mauvaise note:
def indexMini(liste):  
    """détermination de l´indice 
    de la plus mauvaise note"""
    # on commence par initialiser l´indice du minimium au premier élément  
    min=0
    for i in range(len(liste)):
        if liste[i]<liste[min]:
            min=i
    return min
```

```python
# On utilise la méthode pop(i) qui supprime la valeur à l´indice i
# MARTA:
marta.pop(indexMini(marta))
# On affiche la nouvelle liste de notes:
print(marta)
# On affiche la nouvelle moyenne:
print(moyenne (marta))
```

```python
# TIAGO:
tiago.pop(indexMini(tiago))
# On affiche la nouvelle liste de notes:
print(tiago)
# On affiche la nouvelle moyenne:
print(moyenne (tiago))
```

1. On souhaite créer une fonction `total` qui prend un zoo en paramètre ainsi que le nom d’un continent, et qui renvoie le nombre d’animaux originaires de ce continent dans le zoo. Par exemple :<br>
> total (porto, 'Afrique') --> 14 <br>
> total (lisboa, 'Asie') --> 24


a. Quel type de boucle peut-on envisager pour le code de cette fonction ? <br><br>
<span>☐ `for i in dico.keys()`       ☐ `for i in dico.values()`    <br>   ☐ `for (i,j) in dico.items()`       ☐ `Aucune boucle.`</span>


```python
# On itère sur les valeurs qui sont des tuples
# contenant le continent et le nombre d´animaux.
# Exemple à Porto:
for i in porto.values():
    print(i)
```

  b. Écrire et tester le code de cette fonction. 

```python
# fonction
def total(dico,continent):
    # on initialise le nombre d´animaux
    nombre=0
    for i in dico.values():
        # On accède au continent par i[0]
        if i[0]==continent:
            # On accède au nombre par i[1]
            nombre+=i[1]
    return nombre
```

```python
# test de la fonction
print(total (porto, 'Afrique'))
print(total (lisboa, 'Asie'))
```

2. On souhaite créer la fonction `plusGrand` qui prend un zoo en paramètre et qui renvoie le nom de l’animal le plus représenté dans ce zoo. Par exemple :
> plusGrand (porto) --> 'girafe'<br>
> plusGrand (lisboa) --> 'écureuil'


a. Quel type de boucle peut-on envisager pour le code de cette fonction ? <br><br>
<span>☐ `for i in dico.keys()`       ☐ `for i in dico.values()`    <br>   ☐ `for (i,j) in dico.items()`       ☐ `Aucune boucle.`</span>


```python
# On itère sur les items
# i représente les animaux --> les clés
# j représente le tuple (continent,nombre) ---> les valeurs
# Exemple à Lisbone:
for i,j in lisboa.items():
    print(i,j)
```

  b. Écrire et tester le code de cette fonction. 

```python
# fonction: on procède comme la recherche du minimum
# on cherche ici le maximum:
def plusGrand(dico):
    # on initialise le nom et le nombre d´animaux
    animal=''
    nombre=0
    for i,j in dico.items():
        # On accède au nombre par j[1]
        if j[1]>nombre:
            # mise à jour du nombre d´animal:
            nombre=j[1]
            # On accède au nom de l´animal par i:
            animal=i
    return animal
```

```python
# test de la fonction
print(plusGrand(porto))
print(plusGrand(lisboa))
```

  b. Écrire et tester le code de cette fonction. 

```python
# fonction
def nombre(dico,animal):
    # on initialise le nombre d´animaux
    for i,j in dico.items():
        # On accède au nom de l´animal par i:
        if i==animal:
            # on retourne le nombre associé:
            return j[1]
```
